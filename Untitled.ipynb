## Group by and Aggregating

```python
import pandas as pd
```

```python
df = pd.read_csv("/Users/hubertwojcik/Desktop/Projekty/pandas-playground/datasets/Flavors.csv")
df
```

```python
df.groupby('Base Flavor').agg({'Flavor Rating':['mean','max','count','sum']})
```

```python
df.groupby(['Base Flavor','Liked']).agg({'Flavor Rating':[]'})
```
